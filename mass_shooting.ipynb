```python
total = data.Total
```

```python
Q1 = total.quantile(0.25)
Q3 = total.quantile(0.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
seleccion = ((total >= limite_inferior ) & (total <= limite_superior))
data_new = data[seleccion]
```

```python
data_new.boxplot(['Total'])
```
